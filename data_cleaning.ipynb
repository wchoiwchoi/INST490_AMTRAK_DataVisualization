```python
import pandas as pd
import os
```

```python
folder_path = "D:\\School\\2025 Spring\\INST490\\OUR TEAM_S (Cleaned) Copy of Record Reports"
```

```python
# Combines all EXCEL files into a singl DataFrame
all_data = []

for file in os.listdir(folder_path):
    if file.endswith('.xlsx'):
        xls = pd.ExcelFile(os.path.join(folder_path, file))
        for sheet_name in xls.sheet_names:
            df = xls.parse(sheet_name)
            df['source_file'] = file
            df['sheet_name'] = sheet_name
            all_data.append(df)

combined_df = pd.concat(all_data, ignore_index=True)

combined_df
```

```python
#Drop unnecessary columns and rows with missing values
new_df = combined_df[["DATE", "CITY & STATE", "DIVISION ID", "DEPT ID", "ITEMS", "TYPE", "QTY"]]
df_cleaned = new_df.dropna(subset=["DATE"])
```

```python
df_cleaned["CITY & STATE"] = df_cleaned["CITY & STATE"].fillna("")
```

```python
df_cleaned[['CITY', 'STATE']] = df_cleaned['CITY & STATE'].str.split(',', expand=True)

df_cleaned['CITY'] = df_cleaned['CITY'].str.strip()
df_cleaned['STATE'] = df_cleaned['STATE'].str.strip()
df_cleaned['COUNTRY'] = 'United States'

df_cleaned = df_cleaned.drop(columns=["CITY & STATE"])
```

```python
# Fill in missing values in DEPT ID and DIVISION ID columns
df_cleaned["DEPT ID"] = df_cleaned["DEPT ID"].fillna(0)
df_cleaned["DIVISION ID"] = df_cleaned["DIVISION ID"].fillna("")
```

```python
# Drop timestamps from the DATE column
df_cleaned["DATE"] = pd.to_datetime(df_cleaned["DATE"]).dt.date
```

```python
# Convert data types
df_cleaned = df_cleaned.fillna(0).astype({"DATE": str, "CITY": str, "STATE": str, "COUNTRY": str,"DEPT ID" : int, "TYPE" : str, "ITEMS" : str, "QTY": int})
df_cleaned.head()
```

```python
# Capitalize the first letter of each word in TYPE and ITEMS columns
df_cleaned["TYPE"] = df_cleaned["TYPE"].str.capitalize()
df_cleaned["ITEMS"] = df_cleaned["ITEMS"].str.capitalize()
```

```python
# Export the dataframe to an Excel file
df_cleaned.to_excel("total_data.xlsx", index=False)
```
